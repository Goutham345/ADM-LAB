```python
# Generate dataset for customer churn
data_churn = pd.DataFrame({
    'Tenure_Months': np.random.randint(1, 72, 100),
    'Monthly_Charges': np.random.randint(20, 100, 100),
    'Churn': np.random.choice([0, 1], 100)  # 0: No Churn, 1: Churn
})

# Display first few rows
print(data_churn.head())
```

```python
# Define independent variables and target
X_churn = data_churn[['Tenure_Months', 'Monthly_Charges']]
y_churn = data_churn['Churn']
```

```python
# Split data into training and testing sets
X_train_churn, X_test_churn, y_train_churn, y_test_churn = train_test_split(X_churn, y_churn, test_size=0.2, random_state=42)
```

```python
# Create and fit the model
model_churn = LinearRegression()
model_churn.fit(X_train_churn, y_train_churn)

# Print model coefficients
print(f"Churn Model Coefficients: {model_churn.coef_}")
print(f"Churn Model Intercept: {model_churn.intercept_}")
```

```python
# Predict churn probability
y_pred_churn = model_churn.predict(X_test_churn)

# Convert predicted values to binary output (0 or 1 based on 0.5 threshold)
y_pred_churn_binary = [1 if prob > 0.5 else 0 for prob in y_pred_churn]
```

```python
# Calculate model accuracy
accuracy = np.mean(y_pred_churn_binary == y_test_churn)
print(f"Model Accuracy: {accuracy * 100:.2f}%")
```

```python
# Simulated dataset for sales prediction
data_sales = pd.DataFrame({
    'TV': np.random.randint(10, 300, 50),
    'Radio': np.random.randint(5, 50, 50),
    'Newspaper': np.random.randint(1, 20, 50),
    'Sales': np.random.randint(50, 300, 50)
})

# Display first few rows
print(data_sales.head())
```

```python
# Define features and target variable
X_sales = data_sales[['TV', 'Radio', 'Newspaper']]
y_sales = data_sales['Sales']
```

```python
# Split data into training and testing sets
X_train_sales, X_test_sales, y_train_sales, y_test_sales = train_test_split(X_sales, y_sales, test_size=0.2, random_state=42)
```

```python
# Create and train the model
model_sales = LinearRegression()
model_sales.fit(X_train_sales, y_train_sales)

# Print model coefficients
print(f"Sales Model Coefficients: {model_sales.coef_}")
print(f"Sales Model Intercept: {model_sales.intercept_}")
```

```python
# Predict sales for the test set
y_pred_sales = model_sales.predict(X_test_sales)

# Evaluate R-squared value
r2_sales = r2_score(y_test_sales, y_pred_sales)
print(f"R-squared (Sales Model): {r2_sales:.4f}")
```

```python
# Plot Actual vs. Predicted Sales
plt.scatter(y_test_sales, y_pred_sales, color="blue", label="Predicted vs Actual")
plt.xlabel("Actual Sales")
plt.ylabel("Predicted Sales")
plt.title("Actual vs. Predicted Sales")
plt.plot([min(y_test_sales), max(y_test_sales)], [min(y_test_sales), max(y_test_sales)], color="red", linestyle="--", label="Ideal Fit")
plt.legend()
plt.show()
```
